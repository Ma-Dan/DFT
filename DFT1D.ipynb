```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Number of sample points
N = 800
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
#y = np.sin(2 * 2.0*np.pi*x+2.3) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
y = np.zeros(800)

y[:30] = 0
y[30:200] = 2
y[200:] = x[200:]
print(len(y))

plt.plot(x, y, c='b')
```

```python
yf = np.fft.fft(y)
xf = np.linspace(0.0, 1.0//(2.0*T), N//2)

plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
```

```python
frequency_points = 50

xr = np.linspace(0.0, N*T, N)
yr = 0 * xr

for f in range(frequency_points):
    yr = yr + 2.0/N*np.abs(yf[f])*np.cos(f/(N*T)*2.0*np.pi*xr+np.angle(yf[f]))

plt.plot(xr, yr)
```
